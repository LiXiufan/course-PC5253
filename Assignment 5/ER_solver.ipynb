```python
import numpy as np
from scipy.optimize import fsolve

c = 4
phi = 0.4
# f(u) = u - (1 - phi + phi * exp(c * (u - 1)))
func = lambda u: u - (1 - phi + phi * np.exp(c * (u - 1)))

# Search for the non-trivial fixed point non-zero root
u_sol = fsolve(func, 0.5)[0]
S = 1 - np.exp(c * (u_sol - 1))

print(f"u_sol = {u_sol:.3f}")
print(f"S = {S:.3f}")
```
